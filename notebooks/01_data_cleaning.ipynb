```python
df1 = df1.dropna(subset=['Rating', 'Content Rating'])
```

```python
df1 = df1.drop(columns=['Current Ver', 'Android Ver','Last Updated'], errors='ignore')
```

```python
df1['Reviews'] = df1['Reviews'].astype(int)
df1['Installs'] = df1['Installs'].str.replace('+','', regex=False)
df1['Installs'] = df1['Installs'].str.replace(',','')
df1['Installs'] = df1['Installs'].astype(int)
df1['Price'] = df1['Price'].str.replace('$','', regex=False).astype(float)
df1['Type'] = df1['Type'].astype('category')
df1['Category'] = df1['Category'].astype('category')
df1['Content Rating'] = df1['Content Rating'].astype('category')
df1['Genres'] = df1['Genres'].astype('category')
df1['App'] = df1['App'].astype('category')

def clean_size(size):
    if size.endswith('M'):
        return float(size.replace('M',''))
    elif size.endswith('k'):
        return float(size.replace('k','')) / 1024
    else:
        return np.nan  # for "Varies with device"

import numpy as np
df1['Size'] = df1['Size'].apply(clean_size)
df1['Size'] = df1['Size'].fillna(df1['Size'].median())

```

```python
df1.duplicated(subset='App').sum()
```

```python
df1 = df1.drop_duplicates(subset='App')
```

```python
df2 = df2.drop(columns=['Unnamed: 0','id','ver','sup_devices.num','ipadSc_urls.num','vpp_lic','currency'], errors='ignore')
```
